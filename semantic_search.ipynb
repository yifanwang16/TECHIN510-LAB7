```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim

sentences = [
    'The cat is sitting on the mat.',
    'A feline is resting on the rug.',
    'The dog is chasing its tail.',
    'A puppy is running in circles.'
]

model = SentenceTransformer('Supabase/gte-small')
embeddings = model.encode(sentences)
print(cos_sim(embeddings[0], embeddings[1]))
```

```python
# Two lists of sentences
sentences1 = [
    "The sun is shining brightly today.",
    "I love eating ice cream in the summer.",
    "The birds are chirping in the trees.",
]

sentences2 = [
    "It's a beautiful day with clear skies.",
    "Enjoying a cool, refreshing treat on a hot day.",
    "The sound of nature fills the air.",
]

# Compute embedding for both lists
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(sentences2, convert_to_tensor=True)

# Compute cosine-similarities
cosine_scores = cos_sim(embeddings1, embeddings2)

# Output the pairs with their score
for i in range(len(sentences1)):
    print("{} \t\t {} \t\t Score: {:.4f}".format(
        sentences1[i], sentences2[i], cosine_scores[i][i]
    ))
```

```python
# Our sentences to encode
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "I'm looking forward to the weekend.",
    "The best way to predict the future is to create it.",
    "Laughter is the shortest distance between two people."
]
# Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

# Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
```

```python
from sentence_transformers.util import semantic_search

docs = [
    "The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.",
    "The Mona Lisa is a half-length portrait painting by Leonardo da Vinci.",
    "The Statue of Liberty is a colossal neoclassical sculpture on Liberty Island in New York Harbor.",
    "The Colosseum is an oval amphitheatre in the centre of the city of Rome, Italy.",
    "The Great Wall of China is a series of fortifications that were built across the historical northern borders of ancient Chinese states.",
]

docs_embeddings = model.encode(docs, convert_to_tensor=True)

query = "A famous painting"
query_embedding = model.encode(query, convert_to_tensor=True)

hits = semantic_search(query_embedding, docs_embeddings, top_k=3)
hits

for hit in hits[0]:
    print(docs[hit['corpus_id']], "(Score: %.4f)" % hit['score'])
```

```python
import tiktoken

def split_large_text(large_text, max_tokens):
    enc = tiktoken.get_encoding("cl100k_base")
    tokenized_text = enc.encode(large_text)

    chunks = []
    current_chunk = []
    current_length = 0

    for token in tokenized_text:
        current_chunk.append(token)
        current_length += 1

        if current_length >= max_tokens:
            chunks.append(enc.decode(current_chunk).rstrip(' .,;'))
            current_chunk = []
            current_length = 0

    if current_chunk:
        chunks.append(enc.decode(current_chunk).rstrip(' .,;'))

    return chunks
```

Why use tokens?

> By breaking words into smaller parts (tokens), LLMs can better handle new or unusual words by understanding their building blocks. It also helps the model grasp the nuances of language, such as different word forms and contextual meanings.

[source](https://kelvin.legal/understanding-large-language-models-words-versus-tokens/#:~:text=By%20breaking%20words%20into%20smaller,word%20forms%20and%20contextual%20meanings.)

```python
import tiktoken

sent = "The quick brown fox jumps over the lazy dog, and the clever monkey swings from tree to tree in the dense jungle."

print(len(sent.split()))

enc = tiktoken.get_encoding("cl100k_base")
encoded = enc.encode(sent)

print(len(encoded))
tokens = [enc.decode_single_token_bytes(x) for x in encoded]
print(tokens)
print(len(tokens))


decoded = enc.decode(encoded)
print(len(decoded.split()))
decoded
```

```python
doc = """
The quick brown fox jumps over the lazy dog, and the clever monkey swings from tree to tree in the dense jungle.
Meanwhile, a majestic lion roars in the distance, asserting its dominance over the land.
The lush green foliage sways gently in the warm breeze, creating a serene atmosphere.
Colorful birds soar through the sky, their melodic songs echoing through the air.
The jungle is a vibrant ecosystem, teeming with life and endless wonders waiting to be discovered.
"""

split_large_text(doc, 10)
```
